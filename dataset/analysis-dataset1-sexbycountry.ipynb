```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset-1.csv')
```

```python
pd.set_option('display.max_rows', 1500)
```

```python
dfFiltered = df[(df['Subject Area'] == 'Total') &
   (df['First year marker'] == 'All') &
   (df['Mode of study'] == 'All')]
```

```python
pivotDfFiltered = dfFiltered.pivot_table('Number', [
    'Country of HE provider', 'Level of study', 'Sex'], 'Academic Year').reset_index()
```

```python
pivotDfFiltered
```

```python
pivotDfFiltered['% Change 14/15 to 15/16'] = (
    pivotDfFiltered['2015/16'] - pivotDfFiltered['2014/15']) * 100 / pivotDfFiltered['2014/15']
pivotDfFiltered['% Change 15/16 to 16/17'] = (
    pivotDfFiltered['2016/17'] - pivotDfFiltered['2015/16']) * 100 / pivotDfFiltered['2015/16']
pivotDfFiltered['% Change 16/17 to 17/18'] = (
    pivotDfFiltered['2017/18'] - pivotDfFiltered['2016/17']) * 100 / pivotDfFiltered['2016/17']
pivotDfFiltered['% Change 17/18 to 18/19'] = (
    pivotDfFiltered['2018/19'] - pivotDfFiltered['2017/18']) * 100 / pivotDfFiltered['2017/18']
```

```python
pivotDfFiltered
```

## Convert it into 3 datasets for sankey diagram by level of study (All, All undergraduate, All postgraduate) 

### - We won't use row with value of 'Country of HE provider' == 'All', 'Sex' == 'Total' 
### - We use only data in 2018/19

# All (Undergrad+Postgrad)

```python
pivotDfFilteredAll = pivotDfFiltered[
    (pivotDfFiltered['Level of study'] == 'All') &
    ~(pivotDfFiltered['Country of HE provider'] == 'All') &
    ~(pivotDfFiltered['Sex'] == 'Total')
][['Country of HE provider', 'Sex', '2018/19']]
```

```python
pivotDfFilteredAll.reset_index(drop=True, inplace=True)
```

```python
pivotDfFilteredAll = pivotDfFilteredAll.rename(columns={"Academic Year": "",
                                   "Country of HE provider": "source",
                                   "Sex": "target",
                                   "2018/19": "value",
                                  })
```

```python
pivotDfFilteredAll
```

```python
pivotDfFilteredAll.to_csv('sankey-sex-by-country-all.csv', index=False)
```

# Undergraduate

```python
pivotDfFilteredAllUndergrad = pivotDfFiltered[
    (pivotDfFiltered['Level of study'] == 'All undergraduate') &
    ~(pivotDfFiltered['Country of HE provider'] == 'All') &
    ~(pivotDfFiltered['Sex'] == 'Total')
][['Country of HE provider', 'Sex', '2018/19']]
```

```python
pivotDfFilteredAllUndergrad.reset_index(drop=True, inplace=True)
```

```python
pivotDfFilteredAllUndergrad = pivotDfFilteredAllUndergrad.rename(columns={"Academic Year": "",
                                   "Country of HE provider": "source",
                                   "Sex": "target",
                                   "2018/19": "value",
                                  })
```

```python
pivotDfFilteredAllUndergrad
```

```python
pivotDfFilteredAllUndergrad.to_csv('sankey-sex-by-country-all-undergraduate.csv', index=False)
```

# Postgraduate

```python
pivotDfFilteredAllPostgrad = pivotDfFiltered[
    (pivotDfFiltered['Level of study'] == 'All postgraduate') &
    ~(pivotDfFiltered['Country of HE provider'] == 'All') &
    ~(pivotDfFiltered['Sex'] == 'Total')
][['Country of HE provider', 'Sex', '2018/19']]
```

```python
pivotDfFilteredAllPostgrad.reset_index(drop=True, inplace=True)
```

```python
pivotDfFilteredAllPostgrad = pivotDfFilteredAllPostgrad.rename(columns={"Academic Year": "",
                                   "Country of HE provider": "source",
                                   "Sex": "target",
                                   "2018/19": "value",
                                  })
```

```python
pivotDfFilteredAllPostgrad
```

```python
pivotDfFilteredAllPostgrad.to_csv('sankey-sex-by-country-all-postgraduate.csv', index=False)
```

# Multi-level dataset

```python
pivotDfFilteredMulti = pivotDfFilteredAll.copy()
```

```python
pivotDfFilteredMulti = pivotDfFilteredMulti.rename(columns={
                                   "value": "value0",
                                  })
```

```python
pivotDfFilteredMulti['value1'] = pivotDfFilteredAllUndergrad['value']
pivotDfFilteredMulti['value2'] = pivotDfFilteredAllPostgrad['value']
```

```python
pivotDfFilteredMulti
```

```python
pivotDfFilteredMulti.to_csv('sankey-sex-by-country-multi.csv', index=False)
```
