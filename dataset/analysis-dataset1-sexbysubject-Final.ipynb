# Finding average change in proportion of students in each sex over 5-year period, categorised by subject area and level of study

## Started by calculate proportion of students in each sex over 5-year period

```python
dfAllCountries = df[(df['First year marker'] == 'All') &
   (df['Mode of study'] == 'All') &
    (~df['Subject Area'].str.startswith('Total'))]
```

```python
dfAllCountries
```

```python
pivotDfAllCountries = dfAllCountries.pivot_table('Number', ['Subject Area', 'Level of study', 'Country of HE provider', 'Academic Year'], 'Sex').reset_index()
```

```python
pivotDfAllCountries 
```

```python
pivotDfAllCountries[
    (pivotDfAllCountries['Level of study'] == 'Other undergraduate') & 
    (pivotDfAllCountries['Total'] == 0)]
```

```python
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 500)
```

```python
pivotDfAllCountriesAllYear = pivotDfAllCountries.pivot_table(['Female','Male','Other',
                                'Total'], ['Subject Area', 'Level of study', 'Country of HE provider'], 'Academic Year').reset_index()
```

```python
pivotDfAllCountries['Female (%)'] = (pivotDfAllCountries['Female'] / pivotDfAllCountries['Total']) * 100
pivotDfAllCountries['Male (%)'] = (pivotDfAllCountries['Male'] / pivotDfAllCountries['Total']) * 100
pivotDfAllCountries['Other (%)'] = (pivotDfAllCountries['Other'] / pivotDfAllCountries['Total']) * 100
```

```python
pivotDfAllCountries
```

```python
pivotDfAllCountriesAllYear
```

## Then calculate 'average change' in proportion of students in each sex over 5-year period

```python
pivotDfAllCountriesAllYearPercent = pivotDfAllCountriesAllYear.copy()
```

```python
pivotDfAllCountriesAllYearPercent
```

```python
pivotDfAllCountriesAllYearPercent['Female (%) 2014/15'] = (pivotDfAllCountriesAllYearPercent['Female']['2014/15'] / pivotDfAllCountriesAllYearPercent['Total']['2014/15']) * 100
pivotDfAllCountriesAllYearPercent['Male (%) 2014/15'] = (pivotDfAllCountriesAllYearPercent['Male']['2014/15'] / pivotDfAllCountriesAllYearPercent['Total']['2014/15']) * 100
pivotDfAllCountriesAllYearPercent['Other (%) 2014/15'] = (pivotDfAllCountriesAllYearPercent['Other']['2014/15'] / pivotDfAllCountriesAllYearPercent['Total']['2014/15']) * 100
pivotDfAllCountriesAllYearPercent['Female (%) 2015/16'] = (pivotDfAllCountriesAllYearPercent['Female']['2015/16'] / pivotDfAllCountriesAllYearPercent['Total']['2015/16']) * 100
pivotDfAllCountriesAllYearPercent['Male (%) 2015/16'] = (pivotDfAllCountriesAllYearPercent['Male']['2015/16'] / pivotDfAllCountriesAllYearPercent['Total']['2015/16']) * 100
pivotDfAllCountriesAllYearPercent['Other (%) 2015/16'] = (pivotDfAllCountriesAllYearPercent['Other']['2015/16'] / pivotDfAllCountriesAllYearPercent['Total']['2015/16']) * 100
pivotDfAllCountriesAllYearPercent['Female (%) 2016/17'] = (pivotDfAllCountriesAllYearPercent['Female']['2016/17'] / pivotDfAllCountriesAllYearPercent['Total']['2016/17']) * 100
pivotDfAllCountriesAllYearPercent['Male (%) 2016/17'] = (pivotDfAllCountriesAllYearPercent['Male']['2016/17'] / pivotDfAllCountriesAllYearPercent['Total']['2016/17']) * 100
pivotDfAllCountriesAllYearPercent['Other (%) 2016/17'] = (pivotDfAllCountriesAllYearPercent['Other']['2016/17'] / pivotDfAllCountriesAllYearPercent['Total']['2016/17']) * 100
pivotDfAllCountriesAllYearPercent['Female (%) 2017/18'] = (pivotDfAllCountriesAllYearPercent['Female']['2017/18'] / pivotDfAllCountriesAllYearPercent['Total']['2017/18']) * 100
pivotDfAllCountriesAllYearPercent['Male (%) 2017/18'] = (pivotDfAllCountriesAllYearPercent['Male']['2017/18'] / pivotDfAllCountriesAllYearPercent['Total']['2017/18']) * 100
pivotDfAllCountriesAllYearPercent['Other (%) 2017/18'] = (pivotDfAllCountriesAllYearPercent['Other']['2017/18'] / pivotDfAllCountriesAllYearPercent['Total']['2017/18']) * 100
pivotDfAllCountriesAllYearPercent['Female (%) 2018/19'] = (pivotDfAllCountriesAllYearPercent['Female']['2018/19'] / pivotDfAllCountriesAllYearPercent['Total']['2018/19']) * 100
pivotDfAllCountriesAllYearPercent['Male (%) 2018/19'] = (pivotDfAllCountriesAllYearPercent['Male']['2018/19'] / pivotDfAllCountriesAllYearPercent['Total']['2018/19']) * 100
pivotDfAllCountriesAllYearPercent['Other (%) 2018/19'] = (pivotDfAllCountriesAllYearPercent['Other']['2018/19'] / pivotDfAllCountriesAllYearPercent['Total']['2018/19']) * 100
```

```python
pivotDfAllCountriesAllYearPercent
```

```python
pivotDfAllCountriesAllYearPercent = pivotDfAllCountriesAllYearPercent.drop(columns=['Female','Male','Other','Total'])
```

```python
pivotDfAllCountriesAllYearPercent['Avg. % Change of Female'] = ((pivotDfAllCountriesAllYearPercent['Female (%) 2015/16'] - pivotDfAllCountriesAllYearPercent['Female (%) 2014/15']) +
                                                                (pivotDfAllCountriesAllYearPercent['Female (%) 2016/17'] - pivotDfAllCountriesAllYearPercent['Female (%) 2015/16']) +
                                                                (pivotDfAllCountriesAllYearPercent['Female (%) 2017/18'] - pivotDfAllCountriesAllYearPercent['Female (%) 2016/17']) +
                                                                (pivotDfAllCountriesAllYearPercent['Female (%) 2018/19'] - pivotDfAllCountriesAllYearPercent['Female (%) 2017/18'])) / 4
pivotDfAllCountriesAllYearPercent['Avg. % Change of Male'] = ((pivotDfAllCountriesAllYearPercent['Male (%) 2015/16'] - pivotDfAllCountriesAllYearPercent['Male (%) 2014/15']) +
                                                                (pivotDfAllCountriesAllYearPercent['Male (%) 2016/17'] - pivotDfAllCountriesAllYearPercent['Male (%) 2015/16']) +
                                                                (pivotDfAllCountriesAllYearPercent['Male (%) 2017/18'] - pivotDfAllCountriesAllYearPercent['Male (%) 2016/17']) +
                                                                (pivotDfAllCountriesAllYearPercent['Male (%) 2018/19'] - pivotDfAllCountriesAllYearPercent['Male (%) 2017/18'])) / 4
pivotDfAllCountriesAllYearPercent['Avg. % Change of Other'] = ((pivotDfAllCountriesAllYearPercent['Other (%) 2015/16'] - pivotDfAllCountriesAllYearPercent['Other (%) 2014/15']) +
                                                                (pivotDfAllCountriesAllYearPercent['Other (%) 2016/17'] - pivotDfAllCountriesAllYearPercent['Other (%) 2015/16']) +
                                                                (pivotDfAllCountriesAllYearPercent['Other (%) 2017/18'] - pivotDfAllCountriesAllYearPercent['Other (%) 2016/17']) +
                                                                (pivotDfAllCountriesAllYearPercent['Other (%) 2018/19'] - pivotDfAllCountriesAllYearPercent['Other (%) 2017/18'])) / 4
```

```python
pivotDfAllCountriesAllYearAvgPercentChg =  pivotDfAllCountriesAllYearPercent.drop(columns=[
    'Female (%) 2014/15', 'Female (%) 2015/16', 'Female (%) 2016/17', 'Female (%) 2017/18', 'Female (%) 2018/19',
    'Male (%) 2014/15', 'Male (%) 2015/16', 'Male (%) 2016/17', 'Male (%) 2017/18', 'Male (%) 2018/19',
    'Other (%) 2014/15', 'Other (%) 2015/16', 'Other (%) 2016/17', 'Other (%) 2017/18', 'Other (%) 2018/19',
])
```

```python
pivotDfAllCountriesAllYearAvgPercentChg
```

```python
pivotDfAllCountriesAllYearAvgPercentChg.sort_values(by='Avg. % Change of Female', ascending=False)
```

```python
pivotDfAllCountriesAllYearAvgPercentChg[
    (pivotDfAllCountriesAllYearAvgPercentChg['Level of study'] == 'All') &
    (pivotDfAllCountriesAllYearAvgPercentChg['Country of HE provider'] == 'All')
].sort_values(by='Avg. % Change of Female', ascending=False)
```

```python
pivotDfAllCountriesAllYearAvgPercentChg[
    np.isnan(pivotDfAllCountriesAllYearAvgPercentChg['Avg. % Change of Female'])]
```
